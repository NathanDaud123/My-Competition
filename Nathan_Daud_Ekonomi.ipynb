## Proyeksi dan Optimalisasi Ekonomi Kota Banjarmasin Menuju Suksesnya Sustainable Development Goals 2030

### Latar Belakang
Kota Banjarmasin, sebagai ibu kota Provinsi Kalimantan Selatan, memiliki peran strategis dalam perekonomian regional. Dengan tantangan dan peluang yang ada, optimalisasi ekonomi kota ini menjadi krusial untuk mencapai Sustainable Development Goals (SDGs) 2030. SDGs mencakup 17 tujuan global yang meliputi penghapusan kemiskinan, peningkatan kualitas pendidikan, dan pembangunan ekonomi berkelanjutan. Oleh karena itu, proyeksi dan optimalisasi ekonomi Banjarmasin perlu direncanakan secara matang untuk mendukung pencapaian SDGs.

### Tujuan Analisis
1. **Menganalisis tren ekonomi Kota Banjarmasin dari tahun-tahun sebelumnya hingga tahun 2023.**
2. **Membuat proyeksi ekonomi hingga tahun 2030 menggunakan metode regresi linier dan BiLSTM.**
3. **Mengidentifikasi faktor-faktor yang mempengaruhi perekonomian Banjarmasin dan memberikan rekomendasi kebijakan untuk optimalisasi ekonomi.**

### Analisis dan Interpretasi

#### Tren Ekonomi Kota Banjarmasin
Berdasarkan data dari Badan Pusat Statistik (BPS) dan berbagai publikasi, beberapa indikator utama ekonomi Banjarmasin meliputi inflasi, indeks harga konsumen, tingkat pengangguran, dan garis kemiskinan. Tren ini dianalisis menggunakan data historis dari tahun 2000 hingga 2023.

#### Proyeksi Ekonomi Hingga 2030
Proyeksi ekonomi dilakukan menggunakan dua metode:
1. **Regresi Linier:** Metode ini digunakan untuk memprediksi nilai-nilai ekonomi berdasarkan tren historis. Hasilnya menunjukkan proyeksi yang stabil namun linear.
2. **BiLSTM (Bidirectional Long Short Term Memory):** Model RNN ini memberikan proyeksi yang lebih dinamis dan akurat dengan mempertimbangkan pola data yang kompleks. Hasil proyeksi menunjukkan variasi yang lebih realistis dalam indikator ekonomi.

#### Hasil Proyeksi
- **Inflasi:** Diproyeksikan akan mengalami fluktuasi namun cenderung stabil.
- **Indeks Harga Konsumen:** Diperkirakan akan meningkat seiring dengan pertumbuhan ekonomi.
- **Tingkat Pengangguran:** Diproyeksikan menurun dengan adanya kebijakan optimalisasi tenaga kerja.
- **Garis Kemiskinan:** Diperkirakan menurun dengan implementasi program-program pengentasan kemiskinan.

### Kesimpulan
Optimalisasi ekonomi Kota Banjarmasin menuju SDGs 2030 memerlukan pendekatan yang holistik dan terintegrasi. Berdasarkan analisis dan proyeksi yang dilakukan, beberapa rekomendasi kebijakan yang dapat diambil antara lain:
- **Penguatan Infrastruktur:** Meningkatkan ketersediaan dan kualitas infrastruktur dasar untuk mendukung pertumbuhan ekonomi.
- **Pengembangan Sumber Daya Manusia:** Melalui pendidikan dan pelatihan yang sesuai dengan kebutuhan pasar kerja.
- **Peningkatan Investasi:** Mendorong investasi di sektor-sektor strategis seperti industri kreatif dan pariwisata.
- **Pengentasan Kemiskinan:** Implementasi program-program sosial yang efektif untuk mengurangi kemiskinan dan kesenjangan.

Dengan langkah-langkah ini, diharapkan Kota Banjarmasin dapat mencapai target-target SDGs 2030 dan menjadi kota yang lebih berkelanjutan dan sejahtera.

## RNN Buatan --> LSTM dan BiLSTM

```python
# Definisi model LSTM
class LSTM:
    def __init__(self, input_dim, hidden_dim, output_dim):
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim

        # Initialize weights
        self.Wf = np.random.randn(hidden_dim, input_dim + hidden_dim) * 0.01
        self.bf = np.zeros((hidden_dim, 1))
        self.Wi = np.random.randn(hidden_dim, input_dim + hidden_dim) * 0.01
        self.bi = np.zeros((hidden_dim, 1))
        self.Wc = np.random.randn(hidden_dim, input_dim + hidden_dim) * 0.01
        self.bc = np.zeros((hidden_dim, 1))
        self.Wo = np.random.randn(hidden_dim, input_dim + hidden_dim) * 0.01
        self.bo = np.zeros((hidden_dim, 1))
        self.Wy = np.random.randn(output_dim, hidden_dim) * 0.01
        self.by = np.zeros((output_dim, 1))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def tanh(self, x):
        return np.tanh(x)

    def forward(self, x):
        h_prev = np.zeros((self.hidden_dim, 1))
        c_prev = np.zeros((self.hidden_dim, 1))

        for t in range(x.shape[0]):
            x_t = x[t].reshape(-1, 1)
            concat = np.vstack((h_prev, x_t))

            f_t = self.sigmoid(np.dot(self.Wf, concat) + self.bf)
            i_t = self.sigmoid(np.dot(self.Wi, concat) + self.bi)
            c_tilde_t = self.tanh(np.dot(self.Wc, concat) + self.bc)
            c_t = f_t * c_prev + i_t * c_tilde_t
            o_t = self.sigmoid(np.dot(self.Wo, concat) + self.bo)
            h_t = o_t * self.tanh(c_t)

            h_prev = h_t
            c_prev = c_t

        y = np.dot(self.Wy, h_t) + self.by
        return y

# Definisi model biLSTM
class biLSTM:
    def __init__(self, input_dim, hidden_dim, output_dim):
        self.forward_lstm = LSTM(input_dim, hidden_dim, output_dim)
        self.backward_lstm = LSTM(input_dim, hidden_dim, output_dim)

    def forward(self, x):
        forward_output = self.forward_lstm.forward(x)
        backward_output = self.backward_lstm.forward(x[::-1])

        return (forward_output + backward_output) / 2
```

## Implementasi BiLSTM

```python
# Fungsi untuk melakukan proyeksi menggunakan biLSTM
def make_projection(data, target_column, years_to_predict, input_dim, hidden_dim, output_dim):
    sequence_length = len(data)
    x = np.array(data[target_column]).reshape(sequence_length, 1)
    model = biLSTM(input_dim, hidden_dim, output_dim)
    output = model.forward(x)
    last_value = output[-1, 0]

    projections = []
    for i in range(years_to_predict):
        next_value = model.forward(np.array([[last_value]]))[0, 0]
        projections.append(next_value)
        last_value = next_value

    return projections

# Pastikan kolom 'Tahun' adalah tipe datetime
if pd.api.types.is_datetime64_any_dtype(data['Tahun']) is False:
    data['Tahun'] = pd.to_datetime(data['Tahun'], format='%Y')

# Memfilter data dalam rentang tahun 2000 hingga 2023
data = data[(data['Tahun'].dt.year >= 2000) & (data['Tahun'].dt.year <= 2023)]

# Simpan data asli untuk denormalisasi nanti
original_data = data.copy()

# Normalisasi data input
normalized_data = data.copy()
for column in data.columns[1:]:
    normalized_data[column] = (data[column] - data[column].mean()) / data[column].std()

# Proyeksi hingga 2030
years_to_predict = 7
projection_years = np.arange(2024, 2024 + years_to_predict)

# Membuat proyeksi untuk setiap variabel
projection_data = {}
for column in normalized_data.columns[1:]:
    projection_data[column] = make_projection(normalized_data, column, years_to_predict, 1, 5, 1)

# Gabungkan data aktual dan proyeksi
full_data = original_data.copy()
projection_df_dict = {}
for column in normalized_data.columns[1:]:
    projection_df = pd.DataFrame({
        'Tahun': pd.date_range(start='2024', periods=years_to_predict, freq='Y'),
        column: projection_data[column]
    })
    projection_df_dict[column] = projection_df
    full_data = pd.concat([full_data, projection_df], ignore_index=True)

# Denormalisasi data untuk visualisasi
for column in original_data.columns[1:]:
    mean = original_data[column].mean()
    std = original_data[column].std()
    full_data[column] = full_data[column] * std + mean
    projection_df_dict[column][column] = projection_df_dict[column][column] * std + mean

# Membuat subplots untuk semua tren
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(20, 30))

# Visualisasi Tren untuk setiap variabel
variables = original_data.columns[1:]
for i, var in enumerate(variables):
    row, col = divmod(i, 2)
    sns.lineplot(data=full_data, x='Tahun', y=var, ax=axes[row, col], label='Actual')
    sns.lineplot(data=projection_df_dict[var], x='Tahun', y=var, ax=axes[row, col], color='red', linestyle='--', label='Prediction')
    axes[row, col].set_title(f'Tren {var} dari Tahun ke Tahun', fontweight='bold')
    axes[row, col].set_xlabel('Tahun', fontweight='bold')
    axes[row, col].set_ylabel(var, fontweight='bold')
    axes[row, col].legend()

# Menyesuaikan layout
plt.tight_layout()
plt.show()
```

## Analisis Regresi Linier

```python
# Pastikan kolom 'Tahun' adalah tipe datetime
if pd.api.types.is_datetime64_any_dtype(data['Tahun']) is False:
    data['Tahun'] = pd.to_datetime(data['Tahun'].astype(str), format='%Y')

# Memfilter data dalam rentang tahun 2000 hingga 2023
data = data[(data['Tahun'].dt.year >= 2000) & (data['Tahun'].dt.year <= 2023)]

# Simpan data asli untuk denormalisasi nanti
original_data = data.copy()

# Normalisasi data input
normalized_data = data.copy()
for column in data.columns[1:]:
    normalized_data[column] = (data[column] - data[column].mean()) / data[column].std()

# Fungsi untuk melakukan proyeksi menggunakan regresi linier
def make_projection(data, target_column, years_to_predict):
    x = np.array(data['Tahun'].dt.year).reshape(-1, 1)
    y = np.array(data[target_column]).reshape(-1, 1)

    # Menambahkan kolom bias (intercept)
    x = np.hstack((np.ones_like(x), x))

    # Menghitung koefisien regresi linier (beta)
    beta = np.linalg.inv(x.T @ x) @ x.T @ y

    # Membuat proyeksi
    projection_years = np.arange(2024, 2024 + years_to_predict).reshape(-1, 1)
    projection_years = np.hstack((np.ones_like(projection_years), projection_years))
    projections = projection_years @ beta

    return projections.flatten()

# Proyeksi hingga 2030
years_to_predict = 7

# Membuat proyeksi untuk setiap variabel
projection_data = {}
for column in normalized_data.columns[1:]:
    projection_data[column] = make_projection(normalized_data, column, years_to_predict)

# Gabungkan data aktual dan proyeksi
full_data = original_data.copy()
projection_df_dict = {}
for column in normalized_data.columns[1:]:
    projection_df = pd.DataFrame({
        'Tahun': pd.date_range(start='2024', periods=years_to_predict, freq='Y'),
        column: projection_data[column]
    })
    projection_df_dict[column] = projection_df
    full_data = pd.concat([full_data, projection_df], ignore_index=True)

# Denormalisasi data untuk visualisasi
for column in original_data.columns[1:]:
    mean = original_data[column].mean()
    std = original_data[column].std()
    full_data[column] = full_data[column] * std + mean
    projection_df_dict[column][column] = projection_df_dict[column][column] * std + mean

# Membuat subplots untuk semua tren
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(20, 30))

# Visualisasi Tren untuk setiap variabel
variables = original_data.columns[1:]
for i, var in enumerate(variables):
    row, col = divmod(i, 2)
    sns.lineplot(data=full_data, x='Tahun', y=var, ax=axes[row, col], label='Actual')
    sns.lineplot(data=projection_df_dict[var], x='Tahun', y=var, ax=axes[row, col], color='red', linestyle='--', label='Prediction')
    axes[row, col].set_title(f'Tren {var} dari Tahun ke Tahun', fontweight='bold')
    axes[row, col].set_xlabel('Tahun', fontweight='bold')
    axes[row, col].set_ylabel(var, fontweight='bold')
    axes[row, col].legend()

# Menyesuaikan layout
plt.tight_layout()
plt.show()
```

## Implementasi Optimasi dengan Squid Game Optimizer (SGO)

```python
import numpy as np

class SGO:
    def __init__(self, obj_func, num_players, dim, bounds, max_iter):
        self.obj_func = obj_func
        self.num_players = num_players
        self.dim = dim
        self.bounds = bounds
        self.max_iter = max_iter
        self.players = np.random.uniform(bounds[0], bounds[1], (num_players, dim))
        self.best_solution = None
        self.best_fitness = float('inf')

    def evaluate(self, player):
        return self.obj_func(player)

    def handle_boundaries(self, player):
        return np.clip(player, self.bounds[0], self.bounds[1])

    def update_positions(self, offensives, defensives, best_offensive, best_defensive):
        for i in range(len(offensives)):
            r1, r2 = np.random.rand(), np.random.rand()
            offensives[i] += r1 * (best_offensive - offensives[i]) - r2 * (best_defensive - offensives[i])
            offensives[i] = self.handle_boundaries(offensives[i])
        return offensives

    def optimize(self):
        for iteration in range(self.max_iter):
            np.random.shuffle(self.players)
            offensives, defensives = np.split(self.players, 2)
            best_offensive = offensives[np.argmin([self.evaluate(p) for p in offensives])]
            best_defensive = defensives[np.argmin([self.evaluate(p) for p in defensives])]
            offensives = self.update_positions(offensives, defensives, best_offensive, best_defensive)
            self.players = np.vstack((offensives, defensives))
            current_best = min(self.players, key=self.evaluate)
            current_fitness = self.evaluate(current_best)
            if current_fitness < self.best_fitness:
                self.best_fitness = current_fitness
                self.best_solution = current_best
            print(f"Iteration {iteration+1}/{self.max_iter}, Best Fitness: {self.best_fitness}")
        return self.best_solution, self.best_fitness

# Contoh penggunaan
def objective_function(x):
    # Misalkan kita ingin meminimalkan jumlah penduduk miskin berdasarkan inflasi dan indeks harga konsumen
    inflasi, indeks_harga_konsumen = x
    # Fungsi objektif sederhana sebagai contoh
    return inflasi * 0.5 + indeks_harga_konsumen * 0.5

bounds = [1, 10]
num_players = 50
dim = 2
max_iter = 100

sgo = SGO(objective_function, num_players, dim, bounds, max_iter)
best_solution, best_fitness = sgo.optimize()
print("Best Solution:", best_solution)
print("Best Fitness:", best_fitness)
```
